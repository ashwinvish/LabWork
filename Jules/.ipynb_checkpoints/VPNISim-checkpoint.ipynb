```julia
SlopesL = rand(0.5:4.5,10,1) # sensitivity Left side
SlopesR = -1*rand(0.5:4.5,10,1) # sensitivity right side


eyePos = collect(-25:1:25)' # eyeposisitons

eyeThresholdsL = rand(-25:0,10,1)
eyeThresholdsR = rand(0:25,10,1)


ECellsL = zeros(length(eyePos),10) 
ECellsR =  zeros(length(eyePos),10)

yInterceptsL = rand(0:80,10,1) # threshold firing rates
yInterceptsR = -1*rand(0:80,10,1) # threshold firing rates

yInterceptsL = rand(0:20,10,1) # threshold firing rates
yInterceptsR = -1*rand(0:20,10,1) # threshold firing rates


for i =1:1:10
    ECellsL[:,i]  =  max(SlopesL[i,1] * (eyePos[1,:]-eyeThresholdsL[i])  + yInterceptsL[i,1], 0)
    ECellsR[:,i]  =  max(SlopesR[i,1] * (eyePos[1,:]-eyeThresholdsR[i]) - yInterceptsR[i,1],0)
end

# plotting inputs


```

```julia
using Plots

plot(eyePos',ECellsL, color = :blue ,
                        ylim = (0,200),
                        xlim = (-25:25),
                        title = "Firing rate",
                        xlabel = "eye-positon (deg)",
                        ylabel = "Firing rate (Hz)", 
                        lab = "")
                        
plot!(eyePos',abs(ECellsR), color = :red ,
                        lab ="")
```
